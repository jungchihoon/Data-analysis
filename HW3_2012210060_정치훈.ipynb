```python
# We will use only one kind of variable as a time series data
def get_concated_data(raw_data, col_name, lookback):
    x_datas = list()
    y_datas = list()
    # col_name : Variable to be used
    col_data = raw_data[col_name].values
    # aggregate target variable data for length of lookback days
    
    ##### Your Code Here ####
    xd = [[col_data[i:i+lookback]] for i in range(0, len(col_data)-lookback)]
   
    yd = [col_data[i+lookback] for i in range(0, len(col_data)-lookback)]
   ##########################    
    x_data = np.concatenate(xd,0)
    y_data = np.asarray(yd)

    return x_data, y_data
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model
from sklearn.pipeline import make_pipeline

def run_regression(raw_data, column_name, lookback_days, poly_degree):
    # performing regression analysis with specified parameters
    x, y = get_concated_data(raw_data, column_name, lookback_days)
    x_train, x_test, y_train, y_test = split_data(x, y)

    MSE = make_pipeline(PolynomialFeatures(degree = poly_degree), linear_model.LinearRegression(normalize=True))
    MSE.fit(x_train, y_train)
    test_pred = MSE.predict(x_test)
    MSE_test = mean_squared_error(test_pred, y_test)
    MSE_train = mean_squared_error(MSE.predict(x_train), y_train)
    print ("MSE - Train %.4f"%MSE_train)
    print ("MSE - Test %.4f"%MSE_test)
    
    return MSE_train,MSE_test, test_pred, y_test
```

```python
#USE THIS -DO NOT CHANGE-
dist = 1 - cosine_similarity(tfidf_matrix)
MDS()
mds = MDS(n_components=2, dissimilarity="precomputed", random_state=1)
pos = mds.fit_transform(dist)
xs, ys = pos[:, 0], pos[:, 1]

cluster_colors = {0: '#1b9e77', 1: '#d95f02', 2: '#7570b3'}
cluster_names = categories
```

```python
def visualize_clusters(xs, ys, clusters, cluster_names, cluster_colors):
    %matplotlib inline 
    df = pd.DataFrame(dict(x=xs, y=ys, label=clusters)) 
    groups = df.groupby('label')
    fig, ax = plt.subplots(figsize=(17, 9))
    ax.margins(0.05) 
    for name, group in groups:
        ax.plot(group.x, group.y, marker='o', linestyle='', ms=12, 
                label= cluster_names[name],
                color= cluster_colors[name], 
                mec='none')
        ax.set_aspect('auto')
        ax.tick_params(\
            axis= 'x',          
            which='both',      
            bottom='off',      
            top='off',         
            labelbottom='off')
        ax.tick_params(\
            axis= 'y',         
            which='both',     
            left='off',     
            top='off',      
            labelleft='off')
    ax.legend(numpoints=1) 
    plt.show() 

visualize_clusters(xs, ys, clusters, cluster_names, cluster_colors)
```
